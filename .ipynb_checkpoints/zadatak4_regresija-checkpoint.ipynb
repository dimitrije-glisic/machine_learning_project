## Moja implementacija linearne regresije

```python
class MyLinearRegression(object):
    def __init__(self, n_iterations = 30000, learning_rate = 0.03):
        self.n_iterations = n_iterations
        self.learning_rate = learning_rate
        self.n_samples = X.shape[0]

    def initialize_weights(self, n_features):
        limit = 1 / math.sqrt(n_features)
        self.w = np.random.uniform(-limit, limit, (n_features, ))

    def fit(self, X, y):
        X = np.insert(X, 0, 1, axis=1)
        self.training_errors = []
        self.initialize_weights(n_features=X.shape[1])

        for i in range(self.n_iterations):
            y_pred = X.dot(self.w)
            mse = np.mean(0.5 * (y - y_pred)**2)
            self.training_errors.append(mse)
            grad_w = (1/self.n_samples)*(y_pred- y).dot(X)
            self.w -= self.learning_rate * grad_w

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        y_pred = X.dot(self.w)
        return y_pred
```

```python
def r2score(y, ypred):
    ssr = np.sum((y - ypred)**2)
    y_mean = np.mean(y)
    sse = np.sum((y - y_mean)**2)
    r2 = 1 - ssr/sse
    return r2
```

```python
X = df.drop(['cena'], axis=1)
Y = df['cena']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3, random_state=50)   
```

```python
skreg = LinearRegression()
skreg.fit(X_train,Y_train)
ypred = skreg.predict(X_test)
print(f'Sklearn score:{r2score(Y_test, ypred)}')
```

```python
myreg=MyLinearRegression()
myreg.fit(X_train.to_numpy(),Y_train)
ypred = myreg.predict(X_test.to_numpy())
print(f'My score:{r2score(Y_test, ypred)}')
```
