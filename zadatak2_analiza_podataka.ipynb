```python
nekretnine = pd.read_csv('./podaci/data_odbrana', lineterminator='\n', header='infer')
nekretnine.shape
```

```python
def clean_data(row):
    try:
        row.cena = row.cena.replace('€','').replace('.','')
        row.povrsina = row.povrsina.replace('m²', '') 
    except BaseException as e:
        print(row)
    return row
nekretnine = nekretnine.apply(clean_data, axis='columns')

nekretnine[['cena']] = nekretnine[['cena']].astype(float)
nekretnine[['povrsina']] = nekretnine[['povrsina']].astype(float)
```

## f) Prikazati rang listu svih nekretnina izgrađenih u 2020. godini i izlistati ih opadajuće prema ceni prodaje, odnosno ceni iznajmljivanja

```python
import re
def clean_data_godina_izgradnje(row):
    row.godina_izgradnje = re.sub('\D', '', str(row.godina_izgradnje))
    return row

nekretnine = nekretnine.apply(clean_data_godina_izgradnje, axis='columns')
godina_2020 = nekretnine[nekretnine.godina_izgradnje=='2020']
```

### Najskuplji za prodaju

```python
godina_2020[godina_2020.usluga=='prodaja'].sort_values('cena', ascending=False).head(30)
```

### Najskuplji za izdavanje

```python
godina_2020[godina_2020.usluga=='izdavanje'].sort_values('cena', ascending=False).head(30)
```

### g) prikazati nekretnine (Top30) koje imaju:
#### -najveći broj soba unutar nekretnine,
#### -najveću kvadraturu (samo za stanove),
#### -najveću površinu zemljišta (samo za kuće).

```python
def clean_data_broj_soba_plac(row):
    non_decimal = re.compile(r'[^\d.]+')
    row.broj_soba = non_decimal.sub('', str(row.broj_soba))
    row.plac = non_decimal.sub('', str(row.plac))    
    try:
        if row.broj_soba:
            row.broj_soba = float(row.broj_soba)
        else:
            row.broj_soba = 0
        if row.plac:
            row.plac = float(row.plac)
        else:
            row.plac = 0    
    except ValueError as e:
        print(f'could not parse for: {row.broj_soba}')
    return row

nekretnine = nekretnine.apply(clean_data_broj_soba_plac, axis='columns')
```

### Najveci broj soba unutar nekretnine

```python
nekretnine.sort_values('broj_soba', ascending=False).head(30)
```

### Stanovi sa najvecom povrsinom

```python
nekretnine[nekretnine.tip == 'stan'].sort_values('povrsina', ascending = False).head(30)
```

### Kuce sa najvecim placem 

```python
nekretnine[nekretnine.tip == 'kuca'].sort_values('plac', ascending = False).head(30)
```
